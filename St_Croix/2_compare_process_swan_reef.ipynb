```python

import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat              
```

```python
url_restored = r'C:\Users\tkalra\Desktop\NFWP_Swan_coral_USVirgin\69532_Coral_Reefs_St_Croix\Swan_Runs\Restored_Reef\1perc_norefrac\1perc_restored_outs.mat' 
url_restored_dissip = r'C:\Users\tkalra\Desktop\NFWP_Swan_coral_USVirgin\69532_Coral_Reefs_St_Croix\Swan_Runs\Restored_Reef\1perc_norefrac\dissip.mat' 

matfile= loadmat(url_restored)
matfile_dissip = loadmat(url_restored_dissip)    
print(matfile)
xp=matfile["Xp"]
yp=matfile["Yp"]
Hs_restored=matfile["Hsig"]
Setup_restored=matfile["Setup"]
dissip_restored = matfile_dissip["Dissip"]

url_degraded = r'C:\Users\tkalra\Desktop\NFWP_Swan_coral_USVirgin\69532_Coral_Reefs_St_Croix\Swan_Runs\Degraded_Reef\1perc_norefrac\1perc_degraded_outs.mat' 
url_degraded_dissip = r'C:\Users\tkalra\Desktop\NFWP_Swan_coral_USVirgin\69532_Coral_Reefs_St_Croix\Swan_Runs\Degraded_Reef\1perc_norefrac\dissip.mat' 
matfile_dissip = loadmat(url_degraded_dissip)
matfile= loadmat(url_degraded)
    
print(matfile)
xp=matfile["Xp"]
yp=matfile["Yp"]
Hs_degraded=matfile["Hsig"]
Setup_degraded=matfile["Setup"]
dissip_degraded = matfile_dissip["Dissip"]
```

```python
plt.figure(figsize=(10,6))
diff = Hs_restored - Hs_degraded
C = plt.pcolor(xp, yp, diff, cmap='viridis', vmin = -0.1, vmax = 0.1)
cbar = plt.colorbar(C)
plt.title('Difference bet. Hs restored - degraded (meters)')
#plt.savefig('1perc_Restored_Degraded_Hs.png')
print(np.nanmax(diff))
print(np.nanmin(diff))
```

```python
plt.figure(figsize=(10,6))

diff = Setup_restored- Setup_degraded
C = plt.pcolor(xp, yp, diff, cmap='viridis', vmin = -0.1, vmax = 0.1)
cbar = plt.colorbar(C)
plt.title('Difference bet. Setup restored - degraded (meters)')
#plt.savefig('1perc_Restored_Degraded_setup.png')
print(np.nanmin(diff))
print(np.nanmax(diff))
```

```python
plt.figure(figsize=(10,6))
diff = dissip_restored - dissip_degraded
C = plt.pcolor(xp, yp, diff, cmap='viridis', vmin = -0.001, vmax = 0.001)
cbar = plt.colorbar(C)
plt.title('Difference bet. Hs restored - degraded (meters)')
#plt.savefig('1perc_Restored_Degraded_Hs.png')
print(np.nanmax(diff))
print(np.nanmin(diff))
```
