```python

import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import rioxarray
#from scipy.io import loadmat          
```

```python
gridfile = r'C:\Users\tkalra\Desktop\NFWP_Swan_coral_USVirgin\69532_Coral_Reefs_St_Croix\Grid\output_new_small2.tif'
out_grid_file = r'C:\Users\tkalra\Desktop\NFWP_Swan_coral_USVirgin\69532_Coral_Reefs_St_Croix\Grid\restored_USVI_rasterio.tif'
xds = rioxarray.open_rasterio(gridfile, masked=True)
# Hardwired values by manually inspecting the extents of the domain 
bbox = [389822, 192378,   394642, 195148]
bbox_srs = 'EPSG:32163'  # USVI SRS
# Clip the raster to the bounding box
clipped_ds = xds.rio.clip_box(*bbox)#, from_epsg=bbox_srs)

# Save the clipped raster to a new TIFF file
clipped_ds.rio.to_raster(out_grid_file)
```
