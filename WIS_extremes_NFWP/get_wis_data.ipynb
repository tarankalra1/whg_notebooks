```python

import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'ST61022-time_series-20240710T15_42.onlns'
df = pd.read_csv(url,delimiter=' ',header=None)
```

```python
out_file2='wave_timeseries.csv'
df1 = pd.DataFrame()
df1["time"]=(df[0])
df1["Hs"]  = df[10]
df1["Tp"] = df[11]
df1["Wdir"] = df[16]
df1.to_csv("f{out_file2}",index=False)
```

```python
df1
```

```python
# Convert to pandas 
data_time = pd.to_datetime((df1.iloc[:,0]),format='%Y%m%d%H%M%S')
data_time1 = pd.Series(data_time)
```

```python
df2 = pd.DataFrame()
df2["datetime"] = data_time1
df2["Hs"] = df1.iloc[:,1]
df2.set_index('datetime', inplace=True)
df2.to_csv('modified_Hs.csv')
```
