```python
def convert_xy(xx, yy, xx_trans1, yy_trans1):

    xx_trans1.append(xx)
    yy_trans1.append(yy)
     
    xx_trans1 = np.array(xx_trans1)
    yy_trans1 = np.array(yy_trans1)
    
    xx_trans1 = np.array(xx_trans1).ravel()
    yy_trans1 = np.array(yy_trans1).ravel()

    return xx_trans1, yy_trans1
```

```python
cases = ["tidal","100yr_tidal","100yr_rating","500yr_tidal","500yr_rating"]
trans = ["tidal_long_XFBB.dat","100yr_bctide_XFBB.dat","100yr_XFBB.dat","500yr_bctide_XFBB.dat","500yr_XFBB.dat"]

xx_trans1 = []
yy_trans1 = []

xx_trans2 = []
yy_trans2 = []

xx_trans3 = []
yy_trans3 = []

xx_trans4 = []
yy_trans4 = []

xx_trans5 = []
yy_trans5 = []
for i,case in enumerate(cases):  

    url = f'data_plot/historic/{cases[i]}/{trans[i]}'  # Arc id 3 is First Street arc 
    print(url)

# Change the codecs, delim by tabs, no headers
    df = pd.read_csv(url, encoding="utf-16",sep='\t',header=None)

# empty spaces converted to nans
    df=(df.replace(r'^\s*$', np.nan, regex=True))

#convert objects to float 
    df1 = df.astype(float)

    xx=df1[0]
    yy=df1[1]

    if (case=="tidal"):
        print(i)
        print(xx)
        xx_trans1, yy_trans1 = convert_xy(xx, yy, xx_trans1, yy_trans1)

    
    if (case=="100yr_tidal"):
        print(i)
        print(xx)
        xx_trans2, yy_trans2 = convert_xy(xx, yy, xx_trans2, yy_trans2)

    
    if (case=="100yr_rating"):
        print(i)
        print(xx)
        xx_trans3, yy_trans3 = convert_xy(xx, yy, xx_trans3, yy_trans3)

    
    if (case=="500yr_tidal"):
        print(i)
        print(xx)
        xx_trans4, yy_trans4 = convert_xy(xx, yy, xx_trans4, yy_trans4)

    
    if (case=="500yr_rating"):
        print(i)
        print(xx)
        xx_trans5, yy_trans5 = convert_xy(xx, yy, xx_trans5, yy_trans5)
    #yy1 = yy


plt.figure(figsize=(8, 4))
plt.plot(xx_trans1, yy_trans1, 'ro-', label=cases[0])
plt.plot(xx_trans2, yy_trans2, 'k+-', label=cases[1])
plt.plot(xx_trans3, yy_trans3, 'mx-', label=cases[2])
plt.plot(xx_trans4, yy_trans4, 'b', label=cases[3])
plt.plot(xx_trans5, yy_trans5, 'y^-', label=cases[4])
plt.ylabel('velocity mag (ft/s)')
plt.xlabel("cross section across First Street bridge (ft)")
plt.legend()
plt.savefig('xfirst_street.png')
```
